# Basics of Matplotlib and Seaborn - data visualization with Python
---
This notebook teaches the basics of two of the most popular visualization libraries in the Python universe - `Matplotlib` and `Seaborn`. We will show, how using minimal code, you can create various types of plots - 
- Style
- Plot
- Scatter 	
- Pie
- Bar
- Histogram
- Contour
- 3D
- Subplot		
- Annotate
- Imshow
- Legend


## import library 

```python
import matplotlib.pyplot as plt
```

## 1 . Style

```python
a=(1,2,4,6,8,10) # X Label
b=(1,2,3,4,5,6)  # Y label
plt.style.use('classic') # Classic mood
plt.plot(a,b)
```

```python
a=(1,2,4,6,8,10) # X Label
b=(1,2,3,4,5,6)  # Y label
plt.style.use('dark_background')
plt.plot(a,b)
```

```python
a=(1,2,4,6,8,10) # X Label
b=(1,2,3,4,5,6)  # Y label
plt.style.use('seaborn-darkgrid')
plt.plot(a,b)
```

## 2 . Plot


```python
a=(1,2,3,4,5,16,7,8,9,10)
plt.plot(a)
```

```python
a=(1,2,4,6,8,10) # X Label
b=(1,2,3,4,5,6)  # Y label
plt.style.use('ggplot') # Style of graph
plt.plot(a,b)
```

```python
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y)
```

```python
x=(1,2,3,4,5,6,10)
y=(2,4,6,8,10,12,14)
plt.xlabel('time')
plt.ylabel('work')
plt.plot(x,y)
```

```python
x=(1,2,3,4,5,6,10)
y=(2,4,6,8,10,12,14)
plt.title('Time/Work')
plt.xlabel('time')
plt.ylabel('work')
plt.plot(x,y,color='b',marker='o',markersize=11)
```

```python
def p1(x): 
    return x**4 - 4*x**2 + 3*x
def p2(x):  
    return np.sin(10*x)*10

x = np.linspace(-3,3,1000)
plt.plot(x,p1(x),x,p2(x))
```

```python

x=np.arange(0,10,0.005)

plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='-')

plt.plot(x, x + 4, linestyle='dashed')
plt.plot(x, x + 5, linestyle='--')

plt.plot(x, x + 8, linestyle='dashdot')
plt.plot(x, x + 9, linestyle='-.')

plt.plot(x, x + 12, linestyle='dotted')
plt.plot(x, x + 13, linestyle=':')
```
